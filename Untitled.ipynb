```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 


from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score 
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
     confusion_matrix, classification_report,
     roc_auc_score, roc_curve,
     precision_recall_curve, average_precision_score,
     precision_score, recall_score, f1_score
)
import joblib

from imblearn.over_sampling import SMOTE

RND = 42 
np.random.seed(RND)
```

```python
### setting the feature and target
X = df.drop('Class',axis=1)
y = df['Class']

### as its Logestic regression so scalling is important so scalling Amount and time
scaler = StandardScaler()
X[['Amount', 'Time']] = scaler.fit_transform(X[['Amount','Time']])

## save scaler for deployment 
joblib.dump(scaler,"scaler_amount_time.joblib")
```

```python
### TEST TRAIN SPLIT 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state = RND)
### stratiy here is used for the class imbalance as the datset is fully imbalance so startify helps to sets maintain the same proportion of samples for each class as in the original dataset
### in each the training test and validation sets 

print(X_train.shape,X_test.shape)
print("train fraud ratio : ", y_train.mean())
```

```python
### Evlaution function
def eval_classificaton(y_true, y_pred, y_proba=None, prefix="Test"):
    cm = confusion_matrix(y_true, y_pred)
    print(f"------{prefix}---------")
    print("CONFUSION MATRIX")
    print(cm)
    print(classification_report(y_true, y_pred,digits=4))
    if y_proba is not None:
        roc = roc_auc_score(y_true, y_proba)
        auprc = average_precision_score(y_true, y_proba)
        print(f"RPC-AUC: {roc:.4f} , Average Precision (AUPRC) : {auprc:.4f}")
    return cm

# ROC-AUC and AUPRC are both metrics used to evaluate the performance of binary classification models,
# and AUPRC is mostly preffered for the imbalance dataset 
# It measures the trade-off between Precision and Recall for the minority class.
```

```python
###Baseline Logistic Regression (no imbalance handling)
lr = LogisticRegression(solver='liblinear', random_state= RND, max_iter=1000)
### 'l2' (Ridge): Adds a penalty based on the square of the coefficients
# 'l1' (Lasso): Adds a penalty based on the absolute value of the coefficients, promoting sparse solutions (setting some coefficients to zero).
# 'elasticnet': A combination of L1 and L2.
#None: No regularization

## training 
lr.fit(X_train, y_train)

## orediction on the unseen data as the model is trained 
y_pred = lr.predict(X_test)
y_proba = lr.predict_proba(X_test)[:,1]  ## model calculate the probablities score for the test set the [:,1] takes the probalitises of all the class 1 that is 
## fraud which are needed for the metrics like the ROC-AUC and AUPRC 

eval_classificaton(y_test, y_pred, y_proba)

# Small datasets: 'liblinear' is generally a good and efficient choice.
# Large datasets: 'sag' and 'saga' are often faster for larger datasets, particularly when combined with feature scaling.
```

### the model has the summay as 





```python
fpr, tpr , _ = roc_curve(y_test, y_proba)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label = f"AUC={roc_auc_score(y_test, y_proba):.4f}")
plt.plot([0,1],[0,1], '--', alpha=0.5)
plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title("ROC Curve")
plt.legend();plt.show()

# PR
prec, rec, thresh = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(6,5))
plt.plot(rec, prec, label=f"AP={average_precision_score(y_test,y_proba):.4f}")
plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("Precision-Recall Curve")
plt.legend(); plt.show()

###Receiver Operating Characteristic (ROC) Curve and the AUC
```

```python
### IMPROVE IT BEYOND THE BASE LINE FOR THE RECALL and other 

lr_bal = LogisticRegression(class_weight ='balanced' , solver = 'liblinear', random_state=RND, max_iter=1000)
lr_bal.fit(X_train, y_train)

y_pred_bal = lr_bal.predict(X_test)
y_proba_bal = lr_bal.predict_proba(X_test)[:,1]

eval_classificaton(y_test, y_pred_bal, y_proba_bal, prefix = "Test (class_weight='Balanced')")

## the parameter class_weight = balanced makes the proirty of the minority feature class high and putting the panelity on the minority class if missed
##, it prioritizes the minority class by making it "costlier" for the model to get those predictions wrong.
```

###Now when using the paramter balanced the recall has been increased but the precision has decreased 



### SO now we will use the other technique named as SMOTE



```python
#### USING THE SMOTE technique
sm = SMOTE(random_state=RND)
X_res, y_res = sm.fit_resample(X_train, y_train)
print("After SMOTE: ",np.bincount(y_res))

## fir the LR on resampled dataset
lr_sm = LogisticRegression(solver='liblinear', max_iter = 1000, random_state=RND)
lr_sm.fit(X_res, y_res)

y_pred_sm = lr_sm.predict(X_test)
y_proba_sm = lr_sm.predict_proba(X_test)[:,1]

eval_classificaton(y_test, y_pred_sm, y_proba_sm, prefix = "Test (SMOTE trained):")
```

```python
### If we use the LR slef paramter = balanced or the SMOTE technique there is precison and recall tradeoff and the precision is low 
### so we will be using the Threshold Tunning which is an optimize for precision/recall 
```

```python
###THRESHOLD TUNNING------------------------
## first we find the probablites of the smote lr_sm the logistic regression SMOTE previous one used ( and taking the 1 = fraud) 
probs = lr_sm.predict_proba(X_test)[:,1]
precisions, recalls, thresholds = precision_recall_curve(y_test, probs) ### list of precison, recall and thresholds 


### pick the thershold that gives the desired recall or max F1 
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-12)  ## F1 scores formula is applied to each value that we got as a list from the precisions recalls and thresholds 
best_idx = np.nanargmax(f1_scores[:-1]) ## takes the max f1 score from F1 Scores list and stores it as the best idx that will be used as threshold 
best_thresh = thresholds[best_idx] 
print("Best thershold by F1", best_thresh, "F1:", f1_scores[best_idx])


##applyng threshold 
y_pred_thresh = (probs >= best_thresh).astype(int) ##  now here the threshold is set , like of the probs is greater then 
### mean to say that the probabites we gor from the first time, if its grater then the threshold we gor from the F1 scores then it will be considered as fraud(1) 
eval_classificaton(y_test, y_pred_thresh, probs, prefix=f"Test (threshold={best_thresh:.4f})")
```
